```python
import matplotlib.pyplot as plt
import numpy as np

a=[2012,2014,2016,2018,2020]
b=[29,45,32,12,90]
c=[7,8,9,10,11]

width=0.4
p=np.arange(len(a))
p1=p-width/2
p2=p+width/2

plt.bar(p1,b,color="r",label="hello",width=width,align='center')
plt.bar(p2,c,color="y",label="World",width=width,align='center')
#plt.bar(p2,c,color="y",label="World",width=width)

plt.legend()
plt.show()
```
