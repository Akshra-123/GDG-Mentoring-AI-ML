```python


#used a premade model to check how can i have the output in the df form which will include both yearsexperience and salary



import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load the dataset


# Define independent (X) and dependent (y) variables
X = data[['YearsExperience']]
y = data['Salary']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Generate 20 more years of experience values beyond the current dataset
max_experience = X['YearsExperience'].max()
new_experience = np.linspace(max_experience + 0.1, max_experience + 2, 20).reshape(-1, 1)

# Predict salaries for these new experience values
new_salaries = model.predict(new_experience)

# Combine the new experience and predicted salaries into a DataFrame
new_data = pd.DataFrame({
    'YearsExperience': new_experience.flatten(),
    'PredictedSalary': new_salaries
})

print(new_data)
```
