```python
# Basic bar graph program
import matplotlib.pyplot as plt
# here matplotlib is the main library and pyplot is a submodule in matplotlib which provides MATLAB like interface for creating and customising visualisations
x=["python","C","C++","Java"]
y=[12,23,34,54]
plt.bar(x,y)
plt.show()
```

```python
# can give the width of bar plot
# can give colour to bar plots also (whether a single colour to all or different to each)
# can give color to edge also and also can change the width of edge also
import matplotlib.pyplot as plt
x=[2012,2013,2014,2015,2016]
y=[56,67,89,99,45]
# we can add many features like label,title etc.
# can define font-size of label also
plt.xlabel("Year",fontsize=10)
plt.ylabel("Stock prices",fontsize=10)
plt.title("Values of Stock",fontsize=50)
c=["pink","blue","red","yellow","black"]
plt.bar(x,y,width=0.4,color=c,align="center",edgecolor="purple",linewidth=2,linestyle=":",alpha=0.6,label="My Stokes") # there are two parameters in align 1)edge and 2) centre
plt.legend()  # shows the label over the plot that we have defined earlier
# plt.bar(x,y,color="r") if want single color for all
plt.show()
# will get more clarification if first remove the particular operation and then see the difference
# value of alpha parameter can be in between 0 and 1
# label parameter is used to show any label inside the graph and for that we have to call the legend function from plt
''' alpha can take values between 0 and 1:
alpha=0: The element is fully transparent (invisible).
alpha=1: The element is fully opaque (no transparency).
Values between 0 and 1 (like alpha=0.5) give partial transparency, where the element is see-through but still visible. '''
```

```python
# creation of two bar graphs inside one graph only
import matplotlib.pyplot as plt
import numpy as np

a=[2012,2014,2016,2018,2020]
b=[29,45,32,12,90]
c=[7,8,9,10,11]

width=0.4
p=np.arange(len(a))
p1=p-width/2
p2=p+width/2
plt.bar(p1,b,color="r",label="hello",width=width,align='center')
plt.bar(p2,c,color="y",label="World",width=width,align='center')
plt.legend()

plt.show()
# in the output we can see that the x-axis have numbers written and not the words that we provide
```

```python
# we have seen above that the x-axis have numbers written and not the words that we provide
import matplotlib.pyplot as plt
import numpy as np

a=[2012,2014,2016,2018,2020]
b=[29,45,32,12,90]
c=[7,8,9,10,11]

width=0.4
p=np.arange(len(a))
p1=p-width/2
p2=p+width/2
plt.bar(p1,b,color="r",label="hello",width=width,align='edge')
plt.bar(p2,c,color="y",label="World",width=width,align='edge')
plt.legend()
plt.xticks(p+width/2,a)
plt.show()
```

```python
# if I want the x-axis parameter to get tilted somewhat then
import matplotlib.pyplot as plt

import numpy as np

a=["Python","Java","C","C++"]
b=[2,3,4,5]
c=[6,7,8,9]

plt.xlabel("Lang",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.title("lang_preference",fontsize=15)

width=0.2
p=np.arange(len(a))
p1=p-width/2
p2=p+width/2

plt.bar(p1,b,width=width,color='r',label="priority")
plt.bar(p2,c,width=width,color='b',label="necessity")
plt.legend(loc="upper left")
plt.xticks(p+width/2,a,rotation=120)
plt.show()
```

```python
# A scatter plot with various functions
import matplotlib.pyplot as plt
Grades=[1,2,3,4,5]
Students=[5,6,7,8,9]
plt.title("Grading",fontsize=15)
plt.xlabel("Grades",fontsize=10)
plt.ylabel("Students",fontsize=10)
plt.scatter(Grades,Students,color="r",alpha=0.2)
plt.show()
```

```python
# instead of a single color and same size we can also use different colours and sizes
import matplotlib.pyplot as plt
days=[12,23,34,45,56]
attendance=[3,4,1,0,2]
color = ['red','blue','green','yellow','purple']
size=[200,100,150,500,300]
plt.scatter(days,attendance,c=color,s=size,edgecolor='g',linewidth=2)
plt.show()
```

```python
# as we can see we are getting dots if we want any other shape we can do the same also
import matplotlib.pyplot as plt
a=[1,2,3,4,5]
b=[12,3,2,23,12]
size=[200,300,500,190,250]
plt.scatter(a,b,marker="2",s=size)
plt.show()
```

```python
# we can also use cmap property to apply colours
import matplotlib.pyplot as plt
a=[1,2,3,4,5]
b=[12,32,12,43,31]
color=[12,32,42,18,32]
size=[500]
plt.scatter(a,b,c=color,cmap="plasma",s=size,alpha=0.3,edgecolor='r' )
t=plt.colorbar()
t.set_label("Colour Bar",fontsize=10)
plt.show()
```

```python
# we can also print two graphs simultaneously
import matplotlib.pyplot as plt
a=[1,2,3,4,5]
b=[12,31,18,23,21]
c=[43,23,21,53,47]
plt.scatter(a,b,color='r',s=200) # s=size
plt.scatter(a,c,color='b',s=200)
plt.show()
```

```python
import matplotlib.pyplot as plt
a=[1,2,3,4,5]
b=[23,45,12,21,34]
c=[12,3,21,2,31]
plt.xlabel("x-axis",fontsize=20)
plt.ylabel("y-axis",fontsize=20)
plt.title("Line Graph",fontsize=20)
plt.plot(a,b,color='r',linestyle=':',marker='^',label="alpha")
# inside linestyle we can use -- or -
# inside marker we can use s for square o for circle
plt.plot(a,c,color='y',label="beta")
plt.legend(loc='upper center')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
a=np.random.randint(10,80,50)
l=[10,20,30,40,50]
plt.title("Histogram")
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.hist(a)
plt.legend(loc='upper center')
plt.show()
```

```python
# histogram with customisation
import matplotlib.pyplot as plt
import numpy as np
import random
a=np.random.randint(10,80,50)
l=[10,20,30,40,50]
plt.title("Histogram")
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.hist(a, label='new hist every time',color='purple',bins=l,edgecolor='b',cumulative=-1,align='right') # inside cumulative we can use either True or False or -1 or 1)
plt.legend(loc='upper center')
plt.show()
# first remove each paramter and then see the difference
```

```python
# we can also increase the range of numbers on the table using auto parameter as below
import matplotlib.pyplot as plt
import numpy as np
import random
a=np.random.randint(10,50,30)
plt.hist(a,'auto',(0,400),edgecolor='r') # here range will be given as the third parameter only
plt.show()
```

```python
# we can see on y-axis that the numbering is starting from 0 but if we want to change we can use bottom parameter
import matplotlib.pyplot as plt
import numpy as np
import random
a=np.random.randint(10,70,20)
plt.hist(a,bottom=20)
plt.show()
```

```python
# there are various types of histogram that we can make like as described below
# by changing the orientation from vertical to horizontal
import matplotlib.pyplot as plt
import numpy as np
import random
a=np.random.randint(10,60,30)
plt.hist(a,histtype='bar',orientation='horizontal',edgecolor='r') # various types are step,stepfilled,barstacked,bar
plt.show()
```

```python
# wee can also change the width of graphs using rwidth parameter
import matplotlib.pyplot as plt
import numpy as np
import random
a=np.random.randint(10,40,20)
print(a)
c=[35,24,21,23,17,30,18,36,19,15,28,26,28,30,34,30,38,39,19,36]
plt.hist(c,rwidth=0.5)
plt.show()
```

```python
# we can also convert the graph into log table
import matplotlib.pyplot as plt
import numpy as np
import random
a=np.random.randint(10,40,20)
print(a)
c=[35,24,21,23,17,30,18,36,19,15,28,26,28,30,34,30,38,39,19,36]
plt.hist(c,log=True)
plt.axvline(34,color='r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
a=np.random.randint(10,90,10)
plt.pie(a)
plt.show()
```

```python
# we can also give labels to our pie chart as follows
import matplotlib.pyplot as plt
a=[12,34,56,76]
b=["C","C++","Java","Python"]
plt.pie(a,labels=b,labeldistance=0.5) # it is not required to use label distance always with label
plt.show()
```

```python
# we can also do as like getting one part out of the pie chart
import matplotlib.pyplot as plt
a=[12,34,45,67]
b=['hello','namaste','bonjour','sat shri akal']
c=[0.0,0.0,0.1,0.0]
plt.pie(a,explode=c,startangle=180) # first remove the parameter start angle and then see the difference
plt.show()
```

```python
# change in colour + change in font-size
import matplotlib.pyplot as plt
a=[19,30,25]
b=['hello','Namaste','bonjour']
c=['r','y','b','g']
plt.pie(a,colors=c,labels=b,textprops)
plt.show()
```

```python
# we can also see how much percentage is present for a particular color
import matplotlib.pyplot as plt
a=[12,23,34]
plt.pie(a,autopct="%0.0f%%") # 0. jo bhi number likhenge utni digits show hongi after decimal
plt.show()
```

```python
# shadow parameter
import matplotlib.pyplot as plt
a=[34,56,74]
plt.pie(a,shadow=True,colors=['purple','pink','black'])
plt.show()
```

```python
# radius parameter
import matplotlib.pyplot as plt
a=[12,23,34]
plt.pie(a,radius=1.6)
plt.show()
```

```python
# label distance
import matplotlib.pyplot as plt
a=[12,23,34]
b=['C','C++','Python']
plt.pie(a,labels=b,labeldistance=1.1)
plt.show()
```
