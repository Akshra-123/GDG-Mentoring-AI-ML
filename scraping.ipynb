```python
URL = "https://www.makaan.com/price-trends/property-rates-for-buy-in-delhi"
```

```python
from requests import get
import csv
```

```python
r = get(URL)
```

```python
cont = r.text
```

```python
from bs4 import BeautifulSoup
```

```python
bs = BeautifulSoup(cont, 'html.parser')
```

```python
pages = int(bs.select(".pagination")[0].find_all("li")[-2].find('a').text)
```

```python
pages
```

```python
def get_data_from_page(page_content):
    beautiful = BeautifulSoup(page_content, 'html.parser')
    table = beautiful.select('#locality_apartment')[0].find('table')
    items = list(filter(lambda x: ("View" not in x.text and "trend" not in x.text), table.find_all('td')))
    rows = []
    for row_idx in range(int(len(items)/4)):
        base = (4*row_idx)
        rows.append([items[base].text, items[base+1].text, items[base+2].text, items[base+3].text])
    return rows
```

```python
records = []
for page in range(1, pages+1):
    req = get(URL+f'?page={page}')
    records = records + get_data_from_page(req.text)
    
```

```python
records
```

```python
import pandas as pd
pd.DataFrame(records)[1].unique()
```

```python
f = open("delhi.csv", "w")
csvw = csv.writer(f, delimiter=';')
```
