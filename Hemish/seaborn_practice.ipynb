```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.categorical.catplot(iris, x='species', y='sepal_length', kind='swarm')
```

```python
sns.pairplot(iris, hue='species')
```

```python
from sklearn.preprocessing import LabelEncoder
iris['species'] = LabelEncoder().fit_transform(iris['species'])
```

```python
iris
```

```python
sns.heatmap(iris.corr(), annot=True)
```

```python
sns.countplot(x='species', hue="species", data=iris)
```

```python
sns.boxplot(iris, x='species', y='petal_length')
```

```python
sns.displot(iris, x='petal_length', hue='species', kind='kde')
```

```python
sns.ecdfplot(iris, x='petal_width')
```

```python
sns.boxenplot(iris, x='species', y='petal_width')
```

```python
sns.violinplot(iris, x='petal_length', y='petal_width', density_norm='count')
```
